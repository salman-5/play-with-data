```python
X_heart = data_csv.drop('quality', axis=1)
y_heart = data_csv['quality']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# normalize data but only the numerical columns (Age, trestbps, chol, thalach, oldpeak)
X_heart_standardized = X_heart.copy()
X_heart_standardized[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']] = scaler.fit_transform(X_heart[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']])
X_heart_standardized.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
# X_heart_pca = pca.fit_transform(X_heart)
X_heart_pca_std = pca.fit_transform(X_heart_standardized)
print(X_heart_pca_std)
# plot the data
import matplotlib.pyplot as plt
plt.scatter(X_heart_pca_std[:,0], X_heart_pca_std[:,1], c=y_heart, cmap='Spectral')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('With standardization')
plt.savefig("1.png", bbox_inches="tight")
plt.show()

```

```python
# apply UMAP to the data
import umap
X_umap = umap.UMAP(n_components=2, random_state=42).fit_transform(X_heart_pca_std)

# plot the data
plt.scatter(X_umap[:,0], X_umap[:,1], c=y_heart, cmap='Spectral')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.title('With standardization')
plt.savefig("3.png", bbox_inches="tight")
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)

# Apply t-SNE to the data
X_tsne = tsne.fit_transform(X_heart)
X_tsne_standardized = tsne.fit_transform(X_heart_standardized)

plt.scatter(X_tsne_standardized[:,0],
            X_tsne_standardized[:,1], c=y_heart,cmap='Spectral')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.title('With standardization')
plt.savefig("2.png", bbox_inches="tight")
```

```python
from scipy.stats import f_oneway
t_statistic, p_value = f_oneway(data_csv[data_csv['quality'] == 3]['chlorides'],
                                 data_csv[data_csv['quality'] == 4]['chlorides'],
                                 data_csv[data_csv['quality'] == 5]['chlorides'],
                                 data_csv[data_csv['quality'] == 6]['chlorides'],
                                 data_csv[data_csv['quality'] == 7]['chlorides'],
                                 data_csv[data_csv['quality'] == 8]['chlorides'])

print('t-statistic: {}'.format(t_statistic))
print('p-value: {}'.format(p_value))

# check if significant
if p_value < 0.05:
    print('We reject the null hypothesis')
else:
    print('We fail to reject the null hypothesis')

```

```python
def calculate_iou(rect1,rect2):
    x1, y1, w1, h1 = rect1
    x2, y2, w2, h2 = rect2
    inter_width = max(0, min(x1 + w1, x2 + w2) - max(x1, x2))
    inter_height = max(0, min(y1 + h1, y2 + h2) - max(y1, y2))

    inter_area = inter_width * inter_height
    total_area= w1*h1+w2*h2 - inter_area
    iou_score =  inter_area/total_area
    return iou_score

```

```python
iou=[]
for i in range(100):
  iou.append(calculate_iou(grd_rects[i],prd_rects[i]))

pd.DataFrame(iou)

```

```python
import numpy as np
import random
import cv2
from matplotlib import pyplot as plt

import albumentations as A
```

```python
np.random.seed(23391770)
choices = np.random.choice(range(100))
print(choices)

```

```python
def visualize(image, mask, original_image=None, original_mask=None):
    fontsize = 12

    if original_image is None and original_mask is None:
        f, ax = plt.subplots(2, 1, figsize=(8, 8))

        ax[0].imshow(image)
        ax[1].imshow(mask)
    else:
        f, ax = plt.subplots(2, 2, figsize=(8, 8))

        ax[0, 0].imshow(original_image)
        ax[0, 0].set_title('Original image', fontsize=fontsize)

        ax[0, 1].imshow(original_mask)
        ax[0, 1].set_title('Original mask', fontsize=fontsize)

        ax[1,0].imshow(image)
        ax[1, 0].set_title('Rotated image', fontsize=fontsize)

        ax[1, 1].imshow(mask)
        ax[1, 1].set_title('Rotated mask', fontsize=fontsize)
```

```python
image = cv2.imread('{}.png'.format(choices))
mask = cv2.imread('{}_seg.png'.format(choices), cv2.IMREAD_GRAYSCALE)
```

```python
visualize(image, mask)
```

```python
random.seed(23391770)
aug = A.RandomBrightnessContrast(p=1)


augmented = aug(image=image, mask=mask)

image_rot90 = augmented['image']
mask_rot90 = augmented['mask']

visualize(image_rot90, mask_rot90, original_image=image, original_mask=mask)
```

```python
aug = A.Transpose(p=1)

augmented = aug(image=image_rot90, mask=mask_rot90)

image_transposed = augmented['image']
mask_transposed = augmented['mask']

visualize(image_transposed, mask_transposed, original_image=image_rot90, original_mask=mask_rot90)
```

```python
original_height, original_width = image_transposed.shape[:2]
```

```python
aug = A.CenterCrop(p=1, height=original_height, width=original_width)

augmented = aug(image=image_transposed, mask=mask_transposed)

image_center_cropped = augmented['image']
mask_center_cropped = augmented['mask']

print(image_center_cropped.shape, mask_center_cropped.shape)

assert (image_transposed - image_center_cropped).sum() == 0
assert (mask_transposed - mask_center_cropped).sum() == 0

visualize(image_center_cropped, mask_center_cropped, original_image=image_transposed, original_mask=mask_transposed)
```

```python
aug = A.RandomRotate90(p=1)

augmented = aug(image=image_center_cropped, mask=mask_center_cropped)

image_rot90 = augmented['image']
mask_rot90 = augmented['mask']

visualize(image_rot90, mask_rot90, original_image=image, original_mask=mask)
```
